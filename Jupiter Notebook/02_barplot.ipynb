```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
#Load data set
phool=sns.load_dataset("iris")
# size of figure of bar plot
plt.figure(figsize=(4,3))
#draw a barplot
sns.barplot(x="species",y="sepal_width",data=phool)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#load dataset
gihaaz=sns.load_dataset("titanic")
gihaaz
#plot a barplot
sns.barplot(x="sex",y="pclass",hue="alone",data=gihaaz)  #hue defined the colors.
plt.title("another example with changing the values of x and y and using 'hue' plot a barplot")
plt.show()
```

```python
### WE CAN CHANGE ORDER 
# (YOU CAN SHOW FEMALE FIRST AND THEN MALE IN ABOVE GRAPH)
import seaborn as sns
import matplotlib.pyplot as plt
naav=sns.load_dataset("titanic")
naav
sns.barplot(x="sex",y="pclass",hue="alone",data=naav,order=["female","male"])
plt.title("changing the order of above graph 'put female first and then male '")
plt.show()
```

```python
#### YOU CAN CHANGE THE COLORS OF THE GRAPH TOO BY USING COLOR FUNCT.\
import seaborn as sns
import matplotlib.pyplot as plt
naav=sns.load_dataset("titanic")
sns.barplot(x="sex",y="pclass", hue='alone', data=naav, order=["female","male"],color="grey")
plt.title("changing the color of above graph")
plt.show()
```

```python
#### we can remove "error bars" from above barlines (only by adding {confidence intervel(ci)})
import seaborn as sns
import matplotlib.pyplot as plt
naav=sns.load_dataset("titanic")
sns.barplot(x="sex",y="pclass", hue='alone', data=naav, order=["female","male"],color="grey",ci=None)
plt.title("removing error bars by using ci'confidence intervel'")
plt.show()
```

```python
import seaborn as sns
import numpy
import matplotlib.pyplot as plt
naav=sns.load_dataset("titanic")
sns.barplot(x="class",y="fare", hue='sex', data=naav, ci=None,saturation=0.3)
plt.title("changing the saturation of above graph")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy
phool=sns.load_dataset("titanic")
sns.barplot(x="fare", y="class",hue="sex" , data=phool, saturation=0.2)
plt.title("Horizental barplot")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy
phool=sns.load_dataset("titanic")
sns.barplot(x="fare", y="class",data=phool, saturation=0.2,
            linewidth=,facecolor=(1,1,1,1),
            errcolor="2.5",edgecolor="0")
plt.title("Horizental barplot using user defined style of bars")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy
phool=sns.load_dataset("titanic")
sns.barplot(x="fare", y="class",data=phool, saturation=0.2,
            linewidth=4,facecolor=(1,0,0,1),
            errcolor="0.5",edgecolor="0.5")  #edge,error color can be changed from 0 to 1 "0 is darkest & '1' is lighter "
plt.title("Horizental barplot & Analyzing some changes in bars")
plt.show()
```
