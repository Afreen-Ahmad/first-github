# Boxplot

```python
import seaborn as sns
seaborn.set(style="white") # other styles are darkgrid,whitegrid,dark,ticks
naav=seaborn.load_dataset("titanic")

sns.boxplot(x="class",
                y="fare",
                data=naav)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as pd
seaborn.set(style="whitegrid")
tip=sns.load_dataset("tips")
tip
plt.title("My first boxplot") 
 
sns.boxplot(x='day',y='tip', data=tip, saturation=1)#mostly we draw catogorical value on x axis and numaric values on y axis 
```

```python
# we can draw boxplot of a single column
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="white")
tip=sns.load_dataset("tips")
plt.title("boxplot having single value")
sns.boxplot(y=tip["total_bill"])
```

```python
import seaborn as sns
sns.set(style="whitegrid")
tip=sns.load_dataset("tips")
sns.boxplot(x="tip",y="day",hue="smoker",data=tip ,color="#99208f")
#in plotting box "color="you can give any color by copying colorcode on internet"
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
naav= sns.load_dataset("titanic")

sns.boxplot(x="survived", y="age", showmeans=True,
            meanprops={"marker":"*",
                       "markersize":"12",
                       "markeredgecolor":"red"},data=naav,)
# show labels
#plt.xlabel("How many survived"),
#plt.ylabel("Age(years)"),      PANDING=(labels doesn't work)
#plt.title("Box plot kaatiyah gayi dubith ti kaatiyah bachaai")
plt.show()

            
```

# facet_wrap Facet_grid

```python
import seaborn as sns
import matplotlib.pyplot as plt
ex=sns.load_dataset("exercise")
ex.head()
```

```python
ex.tail()
```

# Facet Grids


```python
g=sns.FacetGrid(ex,row='kind',col='time')
g.map(plt.scatter,'pulse','diet')
```

```python
g=sns.FacetGrid(ex,row='time',#col='time'
               hue='kind', hue_order=['rest','walking','running'],
               sharex=False)
g.map(plt.scatter,'pulse','diet')
```

```python
g=sns.FacetGrid(ex,row='time',col='kind')
               #hue='kind', #hue_order=['rest','walking','running
g.map(sns.barplot,'pulse','diet',palette='magma')
```

for box plot

```python
g=sns.FacetGrid(ex,row='time',col='kind')
               #hue='kind', #hue_order=['rest','walking','running
g.map(sns.boxplot,'pulse','diet',palette='magma')
```
